```python
def remove_category_duplicates(categories_dict):
    """Removes any duplicated ingredients from categories_dict."""
    
    already_seen = []
    
    for key in categories_dict.keys():
        counter = -1 #Set to -1 so that index for first value is 0
        for value in categories_dict[key]:
            counter += 1
            if value in already_seen:
                del categories_dict[key][counter] #Delete duplicates from dictionary (no copy)
            else:
                already_seen.append(value)
                
    return categories_dict
```

```python
def types_of_dishes(dishes_types_amount,
                    eligible_types=eligible_types):
    """Asks user for input on the type and amount of dishes they want recommendations for and saves this.
        
        Arguments:
            dishes_type_amount (dict): Empt dict, to be filled with user input
            eligible_types (list): Specifies what types of dishes can be chosen (e.g. weekday)
            
        Returns:
            dishes_type_amount (dict): What kinds of dishes (e.g. weekday) and their amounts
    """
    
    #Print some information to the user
    
    print('You can choose what types of dishes you want recommendations for')
    print('Allowed options: weekend, weekday, starter, dessert')
    print('Separate each entry with a comma (,)')
    dishes_types_amount.clear()
    
    #Request input from user on the types of dishes to be generated
    
    input_check = 0
    while input_check < 1: 
        print('Please enter what type(s) of dishes you want recommendations for')
        input_string = input('').lower()
        input_list = input_string.replace(' ', '').split(',') #Removes whitespace and splits entry into list
        input_count = len(input_list)
        
        eligible_count = len([entry for entry in input_list if entry in eligible_types])
        if eligible_count == input_count: #Nb of valid entries must be same as total entries 
            input_check = 1
        else:
            print('Entry is not valid. Please try again')
            continue
    
    #Create a dictionary that contains dish type-number of dishes pairs
    
    for entry in input_list:
        print('How many', entry, 'dishes do you want to generate?')
        
        check_count = 0
        while check_count < 1:
            try:
                nb_dishes = input('')
                if len(nb_dishes) == 0:
                    print('No information entered. Please enter the number again')
                    continue
                elif int(nb_dishes) == 0:
                    print('Zero is not a valid number')
                    continue
                elif isinstance(int(nb_dishes * 2), int) != True: #Will result in the ValueError below
                    continue
                else:
                    dishes_types_amount[entry] = int(nb_dishes)
                    check_count += 1
            except ValueError:
                print('The number entered is not a numeric integer. Try again')
                continue
    
    return dishes_types_amount
```

```python
def store_selection(store_choice,
                    eligible_stores=eligible_stores):
    """Allows the user to select what store they plan to buy their groceries at."""    
    
    print('What store are you planning to buy your groceries at?')
    print('Stores that you can choose from are (please enter as written below):')
    print(str(eligible_stores).strip('[]')) #Strip [] from list entries for aesthetics
    store_choice.clear()
    
    #Add step to avoid failing on caps
    
    eligible_count = 0
    while eligible_count < 1:
        choice = input('')
        if choice in eligible_stores:
            store_choice.append(choice)
            eligible_count += 1
        else:
            print('That is not a valid store. Please make another choice')
            continue
            
    return store_choice
```

```python
def create_shopping_list(shopping_list,
                         dishes_generated=dishes_generated,
                         recipes_dict=recipes_dict):
    """Creates a shopping list with ingredients from the recipes that were generated.
        
        Arguments:
            shopping_list (list): Empty list, to be filled with ingredients
            dishes_generated (dict): The dishes recommended or selected by the user
            recipes_dict (dict): The masterlist of all recipes, per dish 
            
        Returns:
            shopping_list (list): Everything that needs to be bought at the grocery store
    """
    
    shopping_list.clear()
    
    #Transform dictionary into a list containing only the dishes
    #Not able to do with nested list comprehension - only returns results for the last key
    dish_by_dish = []
    for key in dishes_generated.keys():
        for dish in range(len(dishes_generated[key])):
            dish_by_dish.append(dishes_generated[key][dish])
    
    shopping_list_temp = [recipes_dict[dish] for dish in dish_by_dish]
    
    #Create a flat list with all the ingredients, instead of a nested one
    #Same issue with nested list comprehension as above
    for element in shopping_list_temp:
        for ingredient in range(len(element)):
            shopping_list.append(element[ingredient])
    
    return shopping_list
```

```python
#Function for sorting the shopping list depending on the layout of the store

def sorted_shopping_list(sorted_list,
                         shopping_list=shopping_list, 
                         store_choice=store_choice, 
                         categories_dict=categories_dict, 
                         store_layouts=store_layouts):
    """Sorts the shopping list in the order of the layout of the chosen store.
        
        Arguments:
            sorted_list (list): Empty list that will contain the sorted shopping list
            shopping_list (list): All things needed on the shopping list, but not sorted
            store_choice (list): The store selected by the user, determines the order of categories
            categories_dict (dict): Ingredients w/o measurements in each category (e.g. avocado in Fruits & Vegetables)
            store_layouts (dict): The category layout of each store in scope  
            
        Returns:
            sorted_list (list): Flat shopping list, sorted in the right order
    """
    
    sorted_list.clear()
    category_order = store_layouts[store_choice[0]]
    
    #Create a flat list with ingredients without measurements (generalized ingredient), based on the category list dict
    #Same issue with nested comprehension
    flat_category_list = []
    for sublist in categories_dict.values():
        for item in sublist:
            flat_category_list.append(item)
    
    category_order_dict = {category: [] for category in category_order}
    
    #For each category, match ingredient in shopping list with generalized ingredient and append to dict by category
    for key in categories_dict.keys():
        for ingredient in shopping_list:
            match = difflib.get_close_matches(ingredient, flat_category_list, n=1, cutoff=0.5)
            if match[0] in categories_dict[key]:
                category_order_dict[key].append(ingredient)
            else:
                continue
    
    #Turn dict from step above into a flat, but still sorted, shopping list
    #Same issue with nested comprehension
    for sublist in category_order_dict.values():
        for item in sublist:
            sorted_list.append(item)
                
    return sorted_list
```

```python
import smtplib, ssl # SMTP = Simple Mail Transfer Protocol, SSL = Secure Sockets Layer
# Gmail will encrypt emails using TLS (Transport Layer Security), as this is the more secure successor of SSL

import getpass as gp

from email.mime.text import MIMEText # MIME = Multipurpose Internet Mail Extensions
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication


def send_email(receiver_email, 
               sorted_list=sorted_list):
    """Send an email with the shopping list in it."""
    
    # Set up connection and server settings
    port = 465  # For SSL
    smtp_server = 'smtp.gmail.com'
    context = ssl.create_default_context() # Create a secure SSL context
    # This will load the system’s trusted CA certificates, enable host name checking 
    # and certificate validation, and try to choose reasonably secure protocol and cipher settings

    # Request password for sender email as authentication
    password = gp.getpass('Type your email password and press enter: ')

    # Define some elements of the message
    sender_email = 'food.generator.test@gmail.com'
    message = MIMEMultipart()
    message['Subject'] = 'Här är din inköpslista'
    message['From'] = sender_email
    
    # Create the HTML version of the message
    #body1 = sorted_list
    body2 = """\
    <html>
    <body>
    <p>Hej,</p>
    <p>Här är din inköpslista.</p>
    <p>Hoppas att maten ska smaka bra!</p>
    <br>
    </br>
    </body>
    </html>
    """
    
    # Turn this into an html MIMEText object
    message_text = MIMEText(body2, 'html')
    message.attach(message_text)
    
    # Attach txt file with shopping list in body of the email
    file_attachment = 'C:\\Users\\nystrom jakob\\Documents\\012 ML courses\\Grocery app\\sorted_list.txt'
    message.attach(MIMEText(open(file_attachment).read()))
    
    #When using the with statement, the connection is closed when code block is exited (quit command)
    with smtplib.SMTP_SSL(smtp_server, port, context=context) as server:
        server.login(sender_email, password)
        for receiver in receiver_email:
            message['To'] = receiver
            server.sendmail(sender_email, receiver, message.as_string())
```
